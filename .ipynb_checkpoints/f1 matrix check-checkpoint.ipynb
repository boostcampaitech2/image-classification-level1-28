```python
from sklearn.metrics import confusion_matrix
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def draw_confusion_matrix(target, pred):
        cm = confusion_matrix(target, pred)
#         df = pd.DataFrame(cm/np.sum(cm, axis=1)[:, None], index=list(range(18)), columns=list(range(18)))
        df = pd.DataFrame(cm)
        df = df.fillna(0)  # NaN 값을 0으로 변경

        plt.figure(figsize=(10, 10))
        plt.tight_layout()
        plt.suptitle('Confusion Matrix')
        sn.heatmap(df, annot=True, cmap=sn.color_palette("Blues"))
        plt.xlabel("input2")
        plt.ylabel("input1")
        plt.show("confusion_matrix.png")


def chk_matrix(labels,predict, y_true = np.array([]), y_pred = np.array([])):
        
        y_true = np.concatenate((y_true, labels.cpu().view(-1).numpy()))
        y_pred = np.concatenate((y_pred, predict.cpu().view(-1).numpy()))
        
        
def create_diff_two_df(df1,df2) -> pd.DataFrame:
    
    df1["ans2"] = df2.ans
    new_dataframe = pd.DataFrame()
    new_dataframe = df1[df1.ans != df2.ans]
    return new_dataframe[['ans','ans2']]
```

```python
df1 = pd.read_csv("7716.csv")
df2 = pd.read_csv("7710.csv")
df3 = pd.read_csv("7694.csv")
df4 = pd.read_csv("6847.csv")
```

# ensemble 1,2,3등 비교

```python
diff_df12 = create_diff_two_df(df1,df2) # result column = [ImageID, ans, ans2]
diff_df23 = create_diff_two_df(df2,df3) # result column = [ImageID, ans, ans2]
diff_df13 = create_diff_two_df(df1,df3) # result column = [ImageID, ans, ans2]
```

```python
draw_confusion_matrix(diff_df12.ans,diff_df12.ans2)
draw_confusion_matrix(diff_df23.ans,diff_df23.ans2)
draw_confusion_matrix(diff_df13.ans,diff_df13.ans2)
```
